```python
# Calculate RC (Runs Created)

import pandas as pd
import matplotlib.pyplot as plt

#df = pd.read_csv("lotte_2025_batting.csv")
df = pd.read_csv("kbo_2025_batting.csv")

#display(df.head())
#print(df.columns)

# --- Basic Runs Created ---
df["RC_basic"] = ((df["H"] + df["BB"]) * df["TB"]) / (df["AB"] + df["BB"])

# --- Technical (more complex) Runs Created ---
A = df["H"] + df["BB"] + df["HP"] - df["CS"] - df["GDP"]
B = df["TB"] + 0.26 * (df["BB"] + df["HP"]) + 0.52 * (df["SH"] + df["SF"] + df["SB"])
C = df["AB"] + df["BB"] + df["HP"] + df["SH"] + df["SF"]

df["RC_technical"] = (A * B) / C
df["R_diff"] = df["R"] - df["RC_technical"]

# --- Calculate RC+ ---
print("RC_league", df["RC_technical"][10], df["RC_technical"].sum() - df["RC_technical"][10])
league_RC = df["RC_technical"][10]

# Calculate RC per game for each team
df["RC_per_game"] = df["RC_technical"] / df["G"]

# Calculate league RC per game
league_RC_per_game = league_RC / df["G"][10]

# Create RC+ (relative to league)
df["RC+"] = 100 * (df["RC_per_game"] / league_RC_per_game)

# Round and display
df = df.round({"RC+": 1})

# Calculate diff
df["RC+_diff"] = df["wRC+"] - df["RC+"]


#dftest = (df["RC_technical"] / (df["RC_technical"].sum()- df["RC_technical"][10])) * df["RC+"]
#weighted_RC_plus = dftest.sum() - dftest[10]
#print(weighted_RC_plus) # 100.85136382618539

plt.scatter(df["wRC+"], df["RC+"])

for i, team in enumerate(df["Team"][:-1]):
    plt.text(df["wRC+"][i]+1, df["RC+"][i], team + " " + f"{(df["RC+"][i] / df["wRC+"][i]):.3f}")

plt.xlim(80, 130)  # Set x-axis limits
plt.ylim(80, 130) # Set y-axis limits

plt.xlabel("wRC+")
plt.ylabel("RC+")
plt.title("wRC+ vs RC+")

df = df.sort_values(by='RC+_diff')
display(df[["Team", "RC_per_game", "RC+", "wRC+", "RC+_diff"]])

# Save the updated CSV
#df.to_csv("kbo_2025_with_RC.csv", index=False, encoding="utf-8-sig")
#print(df[["Team", "R", "RC_basic", "RC_technical", "R_diff"]])

# plt.scatter(df["RC_technical"], df["R"])

# for i, team in enumerate(df["Team"][:-1]):
#     plt.text(df["RC_technical"][i]+5, df["R"][i], team + " " + f"{(df["R"][i] / df["RC_technical"][i]):.3f}")

# plt.xlim(550, 800)  # Set x-axis limits
# plt.ylim(550, 800) # Set y-axis limits

# plt.xlabel("Runs Created (technical)")
# plt.ylabel("Actual Runs Scored")
# plt.title("Runs Created vs Actual Runs")

plt.axline((0,0), slope=1, color='gray', linestyle='--')
plt.rcParams['font.family'] = ['Noto Serif KR']  # Set the font family

plt.show()

print("Actual Runs > Runs Created: sequencing hits good")
```

```python
# Calculate Pythagorean Expected Wins

import pandas as pd
import matplotlib.pyplot as plt

dfb = pd.read_csv("kbo_2025_batting.csv")
dfp = pd.read_csv("kbo_2025_pitching.csv")

# Set the exponent (for KBO, 1.83–1.90 works better than MLB's 2.00)
x = 1.90
#MLB: higher scoring consistency, less variance → exponent ≈ 1.9–2.0
#KBO: smaller league, higher variance in run differentials, more bullpen volatility → lower exponent ≈ 1.83–1.90 fits better
#Lower exponent → teams’ win% changes less drastically with small run differences
#Higher exponent → every run differential swings win% more sharply
#In MLB, run differentials tightly predict wins → steeper curve (x ≈ 2.0).
#In KBO, more randomness (bullpen swings, blowouts, park imbalance) → flatter curve (x ≈ 1.8).

# Calculate expected win percentage and expected wins
dfp["Exp_Win%"] = (dfb["R"] ** x) / (dfb["R"] ** x + dfp["R"] ** x)
dfp["Exp_Wins"] = dfp["Exp_Win%"] * dfp["G"]

# Calculate difference between actual and expected wins
dfp["Win_Diff"] = dfp["W"] - dfp["Exp_Wins"]

# Optional: round for readability
dfp = dfp.round({"Exp_Win%": 3, "Exp_Wins": 2, "Win_Diff": 2})

dfp = dfp.sort_values(by='Win_Diff')

# Display the key results
display(dfp[["Team", "W", "Exp_Wins", "Win_Diff"]])

plt.rcParams['font.family'] = ['Noto Serif KR']  # Set the font family

plt.scatter(dfp["Exp_Wins"], dfp["W"])

for i, team in enumerate(df["Team"][:-1]):
    plt.text(dfp["Exp_Wins"][i]+1, dfp["W"][i], team + " " + f"{(dfp["W"][i] / dfp["Exp_Wins"][i]):.3f}")

plt.xlabel("Expected Wins")
plt.ylabel("Wins")
plt.title("Expected Wins vs Actual Wins")
plt.axline((0,0), slope=1, color='gray', linestyle='--')

plt.xlim(40, 100)  # Set x-axis limits
plt.ylim(40, 100) # Set y-axis limits

plt.show()

print("actual wins > expected wins: timing and bullpen leverage good")
```
